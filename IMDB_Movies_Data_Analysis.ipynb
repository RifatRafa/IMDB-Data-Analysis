####  Find all movies that were released between 2006 and 2010, have a rating above 8.0, but made below the 50th percentile in revenue.

```python
cond1 = (imdb_data['Year'] >= 2006) & (imdb_data['Year'] <= 2010)
cond2 = (imdb_data['Rating'] > 8) & (imdb_data['Revenue'] < imdb_data['Revenue'].quantile(0.5))
```

```python
imdb_data[cond1 & cond2]
```

####  Find all movies that were released between 2006 and 2010, have a rating above 8.0, but made below the 25th percentile in revenue.

```python
cond3 = (imdb_data['Rating'] > 8) & (imdb_data['Revenue'] < imdb_data['Revenue'].quantile(0.25))
```

```python
imdb_data[cond1 & cond3]
```

#### Finally give the feedback for all movies with 'Good' if the movie 'rating' is >= 8 and 'Average' if the movie rating doesn't follow the condition given.

```python
def feedback(x):
    if x >= 8:
        return "Good"
    elif ((x < 8) & (x >= 5)):
        return "Average"
    else:
        return "Bellow Average"
```

```python
imdb_data['Feedback'] = imdb_data['Rating'].apply(feedback)
imdb_data.tail(20)
```
